```python
opium_filenames = []
missing_files = []

for fn in tqdm(df_filtered["Etext Number"]):
    full_path = os.path.join(path + "books", str(fn))
    try:
        with open(full_path) as f:
            if 'opium' in f.read():
                opium_filenames.append(fn)
    except FileNotFoundError:
        missing_files.append(fn)

print(f"Found {len(opium_filenames)} opium files, {len(missing_files)} missing files")
```

```python
poppy_filenames = []
missing_files = []

for fn in tqdm(df_filtered["Etext Number"]):
    full_path = os.path.join(path + "books", str(fn))
    try:
        with open(full_path) as f:
            if 'poppy' in f.read() or "poppies" in f.read():
                poppy_filenames.append(fn)
    except FileNotFoundError:
        missing_files.append(fn)

print(f"Found {len(poppy_filenames)} poopy files, {len(missing_files)} missing files")
```

```python
anodyne_filenames = []
missing_files = []

for fn in tqdm(df_filtered["Etext Number"]):
    full_path = os.path.join(path + "books", str(fn))
    try:
        with open(full_path) as f:
            if 'anodyne' in f.read():
                anodyne_filenames.append(fn)
    except FileNotFoundError:
        missing_files.append(fn)

print(f"Found {len(anodyne_filenames)} anodyne files, {len(missing_files)} missing files")
```

```python
opium_derivative_filenames = []
missing_files = []

derivative_names = ["laudanum", "morphine", "codein", "heroin"]

for fn in tqdm(df_filtered["Etext Number"]):
    full_path = os.path.join(path + "books", str(fn))
    try:
        with open(full_path) as f:
            for word in derivative_names:
                if word in f.read():
                    opium_derivative_filenames.append(fn)
    except FileNotFoundError:
        missing_files.append(fn)

print(f"Found {len(opium_derivative_filenames)} files with opium derivative mentions, {len(missing_files)} missing files")
```
