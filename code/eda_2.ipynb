# Filter opium mention

This includes all phrases related to opium as itself, its derivatives, medical terms, and slangs

```python
opium_keywords = [
    'opium', 'poppy', 'anodyne', 'laudanum', 'morphine', 'codein', 'heroin',
    'narcotic', 'soporific', 'paregoric', 'dover\u2019s powder', 'chandu',
    'nepenthe', 'dream-gum', 'dream-stick', 'black smoke'
]
```

Iterate through each 'Etext Number' in `df_filtered`, read the content of each corresponding book, identify all occurrences of the defined `opium_keywords`, and store the 'Etext Number' along with the list of 'Found Keywords' in a new DataFrame.



```python
opium_mentions = []
missing_book_files = []

for etext_number in tqdm(df_filtered["Etext Number"].unique()):
    full_path = os.path.join(path, "books", str(etext_number))
    found_keywords = []
    try:
        with open(full_path, 'r', encoding='utf-8', errors='ignore') as file:
            content = file.read().lower() # Read content and convert to lowercase for case-insensitive matching
            for keyword in opium_keywords:
                if keyword in content:
                    found_keywords.append(keyword)

        if found_keywords: # Only add if at least one keyword is found
            opium_mentions.append({
                'Etext Number': etext_number,
                'Found Keywords': list(set(found_keywords)) # Use set to get unique keywords
            })
    except FileNotFoundError:
        missing_book_files.append(etext_number)
    except Exception as e:
        print(f"Error processing Etext Number {etext_number}: {e}")

opium_mentions_df = pd.DataFrame(opium_mentions)

print(f"\nFound {len(opium_mentions_df)} books with opium-related keywords.")
print(f"Skipped {len(missing_book_files)} books due to missing files.")
print("\nHead of the new DataFrame:")
display(opium_mentions_df.head())
```

```python
# Merge with the metadata earlier
df_merged = pd.merge(df_filtered, opium_mentions_df, on='Etext Number', how='right')

print("Shape of the merged DataFrame:", df_merged.shape)
print("\nHead of the merged DataFrame:")
display(df_merged.head())
```

```python
# look for authors name using Wikipedia's API and search endpoint
import html
import re
import unicodedata
from functools import lru_cache

import requests

WIKI_API_URL = "https://en.wikipedia.org/w/api.php"
WIKI_TIMEOUT = 15
AUTHOR_HINTS = (
    "author",
    "writer",
    "novelist",
    "poet",
    "playwright",
    "journalist",
    "essayist",
    "dramatist",
    "historian",
    "philosopher",
    "critic",
    "translator",
)

wiki_session = requests.Session()
wiki_session.headers.update({
    "User-Agent": "opium-gutenberg/1.0 (research notebook)"
})

def normalize_author_name(author_name):
    return re.sub(r"\s+", " ", str(author_name)).strip(" ,")

def ascii_fold(text):
    return unicodedata.normalize("NFKD", text).encode("ascii", "ignore").decode("ascii")

def normalize_for_match(text):
    text = ascii_fold(text).lower()
    text = re.sub(r"[^a-z0-9\s]", " ", text)
    return re.sub(r"\s+", " ", text).strip()

def unique_non_empty(values):
    seen = set()
    ordered_values = []
    for value in values:
        cleaned_value = normalize_author_name(value)
        if cleaned_value and cleaned_value not in seen:
            seen.add(cleaned_value)
            ordered_values.append(cleaned_value)
    return ordered_values

def generate_author_candidates(author_name):
    base_name = normalize_author_name(author_name)
    stripped_name = re.sub(r"\s*\([^)]*\)", "", base_name).strip()

    candidates = [
        base_name,
        base_name.replace(".", ""),
        stripped_name,
        stripped_name.replace(".", ""),
    ]

    if "," in base_name:
        parts = [part.strip() for part in base_name.split(",") if part.strip()]
        if len(parts) >= 2:
            reordered_name = " ".join(parts[1:] + [parts[0]])
            candidates.extend([
                reordered_name,
                reordered_name.replace(".", ""),
                re.sub(r"\s*\([^)]*\)", "", reordered_name).strip(),
            ])

    ascii_candidates = [ascii_fold(candidate) for candidate in candidates if candidate]
    candidates.extend(ascii_candidates)
    candidates.extend([candidate.replace(".", "") for candidate in ascii_candidates])

    return unique_non_empty(candidates)

def wikipedia_request(**params):
    response = wiki_session.get(
        WIKI_API_URL,
        params={"format": "json", "utf8": 1, **params},
        timeout=WIKI_TIMEOUT,
    )
    response.raise_for_status()
    return response.json()

@lru_cache(maxsize=2048)
def fetch_page_by_title(title):
    try:
        payload = wikipedia_request(
            action="query",
            redirects=1,
            prop="extracts|info|description",
            exintro=1,
            explaintext=1,
            inprop="url",
            titles=title,
        )
    except requests.RequestException:
        return None

    pages = payload.get("query", {}).get("pages", {})
    for page in pages.values():
        if "missing" in page:
            continue
        return {
            "title": page.get("title", title),
            "summary": page.get("extract") or None,
            "description": page.get("description") or None,
            "url": page.get("fullurl") or None,
        }

    return None

def score_search_result(candidate, result):
    candidate_normalized = normalize_for_match(candidate)
    title = html.unescape(result.get("title", ""))
    snippet = html.unescape(re.sub(r"<[^>]+>", " ", result.get("snippet", "")))

    title_normalized = normalize_for_match(title)
    snippet_normalized = normalize_for_match(snippet)
    candidate_tokens = set(candidate_normalized.split())
    title_tokens = set(title_normalized.split())

    score = 0
    if title_normalized == candidate_normalized:
        score += 100
    if candidate_normalized and title_normalized.startswith(candidate_normalized):
        score += 30
    if candidate_tokens and candidate_tokens.issubset(title_tokens):
        score += 20
    score += 5 * len(candidate_tokens & title_tokens)

    if any(hint in snippet_normalized for hint in AUTHOR_HINTS):
        score += 15
    if "disambiguation" in title_normalized or "disambiguation" in snippet_normalized:
        score -= 50

    return score

@lru_cache(maxsize=2048)
def search_wikipedia_title(candidate):
    try:
        payload = wikipedia_request(
            action="query",
            list="search",
            srsearch=candidate,
            srlimit=10,
            srprop="snippet",
        )
    except requests.RequestException:
        return None

    results = payload.get("query", {}).get("search", [])
    if not results:
        return None

    scored_results = [
        (score_search_result(candidate, result), result)
        for result in results
    ]
    best_score, best_result = max(scored_results, key=lambda item: item[0])
    if best_score < 20:
        return None

    return best_result.get("title")

def get_wikipedia_info(author_name):
    candidates = generate_author_candidates(author_name)

    for candidate in candidates:
        page = fetch_page_by_title(candidate)
        if page:
            return {**page, "lookup_name": candidate}

    for candidate in candidates:
        resolved_title = search_wikipedia_title(candidate)
        if not resolved_title:
            continue
        page = fetch_page_by_title(resolved_title)
        if page:
            return {**page, "lookup_name": candidate}

    return None

author_info = []
for author in tqdm(authors_and_counts.index):
    info = get_wikipedia_info(author)
    if info:
        author_info.append({
            "author": author,
            "lookup_name": info["lookup_name"],
            "title": info["title"],
            "description": info["description"],
            "summary": info["summary"],
            "url": info["url"],
        })

author_info_df = pd.DataFrame(author_info)
author_info_df
```

```python
import re

NATIONALITY_PATTERN = re.compile(
    r"\b("
    r"American|Argentine|Australian|Austrian|Belgian|Brazilian|British|Canadian|Chilean|Chinese|Colombian|"
    r"Croatian|Cuban|Czech|Danish|Dutch|Egyptian|English|Finnish|French|German|Greek|Hungarian|"
    r"Indian|Irish|Israeli|Italian|Japanese|Lebanese|Mexican|New Zealand|Norwegian|Persian|Polish|"
    r"Portuguese|Romanian|Russian|Scottish|Serbian|Slovenian|South African|Spanish|Swedish|Swiss|"
    r"Syrian|Turkish|Ukrainian|Welsh"
    r")\b"
 )

def extract_author_info(author_name, summary, description=None):
    info = {
        "author": author_name,
        "nationality": None,
        "birth_date": None,
        "death_date": None,
        "gender": None,
    }

    text_parts = [part for part in [description, summary] if isinstance(part, str) and part]
    if not text_parts:
        return info

    text = " ".join(text_parts)
    text_lower = f" {text.lower()} "

    if any(pronoun in text_lower for pronoun in [" she ", " her ", " hers "]):
        info["gender"] = "Female"
    elif any(pronoun in text_lower for pronoun in [" he ", " him ", " his "]):
        info["gender"] = "Male"

    full_date_pattern = r"(\d{1,2}\s+\w+\s+\d{4})\s*(?:–|-|to)\s*(\d{1,2}\s+\w+\s+\d{4})"
    year_range_pattern = r"\b(\d{4})\s*(?:–|-|to)\s*(\d{4})\b"
    born_pattern = r"\bborn[^\d]*(\d{1,2}\s+\w+\s+\d{4}|\d{4})"
    died_pattern = r"\bdied[^\d]*(\d{1,2}\s+\w+\s+\d{4}|\d{4})"

    date_match = re.search(full_date_pattern, text)
    if date_match:
        info["birth_date"] = date_match.group(1)
        info["death_date"] = date_match.group(2)
    else:
        year_match = re.search(year_range_pattern, text)
        if year_match:
            info["birth_date"] = year_match.group(1)
            info["death_date"] = year_match.group(2)

    born_match = re.search(born_pattern, text, re.IGNORECASE)
    if born_match and info["birth_date"] is None:
        info["birth_date"] = born_match.group(1)

    died_match = re.search(died_pattern, text, re.IGNORECASE)
    if died_match and info["death_date"] is None:
        info["death_date"] = died_match.group(1)

    nationality_match = NATIONALITY_PATTERN.search(text)
    if nationality_match:
        info["nationality"] = nationality_match.group(1)

    return info

author_details = []
for idx, row in author_info_df.iterrows():
    details = extract_author_info(
        row["author"],
        row["summary"],
        row.get("description"),
    )
    author_details.append(details)

author_details_df = pd.DataFrame(author_details)
author_details_df
```
